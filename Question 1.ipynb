### (a)

The PMF is: $ P(X = x) = \frac{1}{6}, \quad x \in \{1,2,3,4,5,6\}$

The CDF is: $ P(X \leq k) = \sum_{i=1}^k P(X=i) $

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 7)
pmf = np.ones(6)/6
cdf = np.cumsum(pmf)

#PMF
plt.scatter(x, pmf)
plt.title("PMF of a fair die")
plt.xlabel("x")
plt.ylabel("P(X = x)")
plt.show()

#CDF
plt.step(x, cdf, where = "post")
plt.title("CDF of a fair die")
plt.xlabel("x")
plt.ylabel("P(X <= x)")
plt.ylim = (0, 1.1)
plt.show()
```

$$ \mathbb{E}[X] = \sum_{k=1}^6 k\times P(X=k) = (1+2+3+4+5+6)\times \frac{1}{6} = \frac{7}{2}$$

$$ \mathbb{E}[X^2] = \sum_{k=1}^6 k^2\times P(X=k) = (1+4+9+16+25+36)\times \frac{1}{6} = \frac{91}{6}$$

$$Var(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2 = \frac{91}{6} - (\frac{7}{2})^2 = \frac{35}{12} $$
